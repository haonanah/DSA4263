## Data preprocessing

### Remove duplicates

```python
print(trans_df.shape)
```

```python
trans_df_final = trans_df.drop_duplicates()
```

```python
print(trans_df_final.shape)
```

### Convert event_timestamp to datetime

```python
trans_df_final['EVENT_TIMESTAMP'] = pd.to_datetime(trans_df_final['EVENT_TIMESTAMP'])
```

### Check for missing values

### Convert categorical to numerical

```python
categorical_columns = ['billing_state', 'customer_job', 'product_category', 'payment_currency', 'merchant']
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    trans_df_final[col] = le.fit_transform(trans_df_final[col])
    label_encoders[col] = le
```

### Extract domain from email

```python
trans_df_final['email_domain'] = trans_df_final['customer_email'].apply(lambda x: urlparse(f"http://{x.split('@')[-1]}").netloc if pd.notnull(x) else 'unknown')
```

### Extract device type from user agent

```python
trans_df_final['is_mobile'] = trans_df_final['user_agent'].str.contains("Mobile|iPhone|Android", na=False).astype(int)
```

### Drop PII information and irrelevant columns

```python
trans_df_final.drop(['LABEL_TIMESTAMP','customer_name', 'billing_phone', 'customer_email', 'user_agent', 'ENTITY_TYPE'], axis=1, inplace=True)
```
